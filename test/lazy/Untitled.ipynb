```python
import pandas as pd
```

```python
train = pd.read_csv('http://bit.ly/kaggletrain')
```

```python
train.loc[:, ['Name', 'Sex']]
```

```python
train.fillna(0)
```

```python
y = train.Survived
```

```python
X = train.loc[:, ['Sex', 'Age', 'Parch', 'Pclass', 'SibSp']]
X.Sex.replace({'female': 0, 'male': 1}, inplace=True)
X.fillna(0, inplace=True)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
clf = RandomForestClassifier(max_depth=8)
clf.fit(X.loc[:691, :], y[:692])
```

```python
train_preds = clf.predict(X.loc[:691, :])
test_preds = clf.predict(X.loc[691:791, :])
```

```python
import numpy as np
test_error = np.mean(np.abs(y.data[691:792] - test_preds))
train_error = np.mean(np.abs(y.data[:692] - train_preds))
```
